**1. Create a program to create two class.**

**1.1.** Create a constructor and a method for each class                           
**1.2.** Create a __init__.py for adding all packages                           
**1.3.** Import the respective packages                           
**1.4.** Call each class by creating an object to it                            
**1.5.** Create a program by all the above

```python
class program1:
 
    # default constructor
    def __init__(self):
        self.met = "program1"
 
    # a method for printing data members
    def print_met(self):
        print(self.met)
class program2:
 
    # default constructor
    def __init__(self):
        self.met = "program2"
 
    # a method for printing data members
    def print_met(self):
        print(self.met)
 
# creating object of the class
obj1 = program1()
obj2 = program2()
# calling the instance method using the object obj
obj1.print_met()
obj2.print_met()
```
