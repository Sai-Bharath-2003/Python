**1.**Different ways creating a string


```python
# defining strings in Python
my_string = 'Hello'
print(my_string)

my_string = "Hello"
print(my_string)

my_string = '''Hello'''
print(my_string)

# triple quotes string can extend multiple lines
my_string = """Hello, welcome to
           the world of Python"""
print(my_string)
```

**2.**Concatenating two strings using + operator

```python
str1="Hello"
str2="World"
print ("String 1:",str1)
print ("String 2:",str2)
str1=str1*3+str2
print("Concatenated same string:",str1)
```

**3.**Finding the length of the string

```python
stir = "Hello World"
print(len(stir))
```

**4.**Extract a string using Substring

```python
stir = "Hello World"
sub_str=stir[0:5]
print(sub_str)
```

**5.**Searching in strings using index()

```python
text = 'Python is fun'

# find the index of is
result = text.index('is')
print(result)
```

**6.**Matching a String Against a Regular Expression With matches()


```python
import re
#
search_string = "TutorialsPoint"
pattern = "Tutorials"
match = re.match(pattern, search_string)
#If-statement after search() tests if it succeeded
if match:
   print("regex matches: ", match.group())
else:
   print('pattern not found')
```

**7.**Comparing strings

```python

print("Hello" == "hello")
print("Hello" < "hello")
print("Hello" > "Hello")
print("Hello" != "Hello")
```

**8.**startsWith(), endsWith() and compareTo()

```python
stir = "Hello World"
  
# startswith()
print(stir.startswith("Hello"))

# endswith
print(stir.endswith("World"))

#compareTo()
print(stir == 'Hello')
```

**9.**Trimming strings with strip()

```python
inp_str = "Sai Bharath@@@"
print("Input String:")
print(inp_str)
res = inp_str.strip('@')
print("\nString after trimming extra leading and trailing spaces:")
print(res)
```

**10.**Replacing characters in strings with replace()

```python
stir = "Hellllo World"
print(stir.replace("Hellllo", "Hello"))
```

**11.**Splitting strings with split()


```python
text = 'Hello World'
# Splits at space
print(text.split())
```
